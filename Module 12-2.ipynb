```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('loan.csv', low_memory=False)
```

## Question 1A


```python
sns.set(style='whitegrid')
g = sns.catplot(y='annual_inc', data=df, kind='bar')
g.set_ylabels('Income')
plt.show()
```

```python
g = sns.catplot(y='annual_inc', kind='point',ci=95, data=df)
g.set_ylabels('Incomes')
plt.show()
```

```python
g = sns.boxplot(y='annual_inc', data=df)
plt.show()
```

```python
sns.distplot(df.annual_inc, bins=10, kde=False)
plt.show()
```

## Question 1B


```python
g = sns.lmplot(y='int_rate', 
               x='dti',
               data=df, 
               fit_reg=False, 
               scatter_kws={'alpha':0.2}) 
g.set_ylabels("Interest Rate")
g.set_xlabels("DTI")
plt.xlim(0,100)
plt.title('Scatterplot: Interest Rate vs DTI')
plt.show()
```
