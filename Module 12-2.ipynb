```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('loan.csv', low_memory=False)
df1000 = df.sample(1000)
```

## Question 1A


```python
sns.set(style='whitegrid')
g = sns.catplot(y='annual_inc', data=df1000, kind='bar')
g.set_ylabels('Income')
plt.show()
```

```python
g = sns.catplot(y='annual_inc', kind='point',ci=98, data=df1000)
g.set_ylabels('Incomes')
plt.show()
```

```python
g = sns.boxplot(y='annual_inc', data=df1000)
plt.show()
```

```python
sns.distplot(df1000.annual_inc, bins=[0, 20000, 40000, 60000, 80000, 100000, 150000, 200000], kde=False)
plt.show()
```

## Question 1B


```python
gg = sns.jointplot(y='all_util', 
               x='dti',
               data=df1000, 
               kind='hex',
               dropna=True,
               xlim=(0,40),
               ylim=(0,100))
plt.show()
```

```python
gg = sns.jointplot(y='all_util', 
               x='dti',
               data=df1000, 
               kind='kde',
               xlim=(0,40),
               ylim=(0,100))
plt.show()
```

```python
g = sns.lmplot(x='dti', y='all_util',
                data=df1000, scatter_kws={'alpha':0.2}, fit_reg=False)
plt.show()
```

```python
g = sns.stripplot(x='grade', y='all_util', data=df1000)
plt.show()
```

```python
g = sns.stripplot(x='all_util', y='grade', data=df1000)
plt.show()
```

```python
g = sns.boxplot(x="grade", y="all_util", data=df1000)
```

```python
g = sns.swarmplot(x="grade", y="all_util", data=df1000)
plt.show()
```

```python
g = sns.pointplot(x="grade", y="all_util", data=df1000, capsize=.2, hue='grade')
plt.show()
```

```python
g = sns.barplot(x="grade", y="all_util", data=df1000)
plt.show()
```
