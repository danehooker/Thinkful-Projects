```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('purchases.csv')
print(df)
```

```python
df['purchase_per_view'] = df['items_purchased'] / df['ad_views']
```

```python
print(df)
```

```python
np.random.seed(100)
ran = []
x = 0
while x < 10:
    ran.append(np.random.randint(0,100))
    x = x+1
print(ran)
```

```python
df['random_number'] = ran
print(df)
```

```python
plt.plot(df['ad_views'])
plt.ylabel('Ads Viewed')
plt.show()
```

```python
plt.plot(df['ad_views'], color='red')
plt.plot(df['random_number'], color='green')
plt.show()
```

```python
plt.scatter(x=df['ad_views'], y=df['items_purchased'])
plt.show()
```

```python
ranlist = np.random.normal(10, 5, 1000)
```

```python
plt.subplot(2,2,1)
plt.plot(df['random_number'], color='black')
plt.title('Random Series')

plt.subplot(2,2,2)
plt.scatter(x=df['ad_views'], y=df['items_purchased'])
plt.title('Items Purchased vs Ads Viewed')

plt.subplot(2,2,3)
plt.hist(ranlist)

plt.subplot(2,2,4)
plt.boxplot(ranlist)

plt.show()

```
