```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
raw_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv')
```

```python
raw_data.head()
```

```python
raw_data['high_score'] = raw_data.rating > 3.999
```

```python
raw_data.head()
```

```python
list(raw_data)
```

```python
mylist = ['22-minute meals', '3-ingredient recipes', 'apple', 'avocado', 'bacon', 'banana', 'beef', 'bread', 'breakfast', 'broccoli', 'butter', 'candy', 'cheese', 'chocolate', 'christmas', 'cookies', 'dairy free', 'dessert', 'drinks', 'fish', 'fry', 'healthy', 'low carb', 'low fat', 'low sugar', 'pasta', 'salad', 'seafood', 'soup/stew', 'vegetarian', 'high_score']
```

```python
len(mylist)
```

```python
df = raw_data[mylist]
```

```python
df.head()
```

```python
svr = SVR()
X = df.drop(['high_score'], 1)
Y = df['high_score']
svr.fit(X,Y)
```

```python
svr.score(X, Y)
```

```python
cross_val_score(svr, X, Y, cv=5)
```
